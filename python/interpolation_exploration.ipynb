```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt(pow(x1-x2, 2) + pow(y1-y2, 2))

def score(x, y):
    d_tl = 1 / (euclidean_distance(x, y, 1.0, 0.0) + 1e-3)
    d_tr = 1 / (euclidean_distance(x, y, 1.0, 1.0) + 1e-3)
    d_bl = 1 / (euclidean_distance(x, y, 0.0, 0.0) + 1e-3)
    d_br = 1 / (euclidean_distance(x, y, 0.0, 1.0) + 1e-3)
    return (4 * d_tr + 3 * d_tl + 2 * d_br + 1 * d_bl) / np.sum([d_tl, d_tr, d_bl, d_br])

X = np.linspace(0.0, 1.0, num=501)
Y = np.linspace(0.0, 1.0, num=501)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

Z = np.zeros((501, 501))

for x in X: 
    for y in Y: 
        Z[int(x*500)][int(y*500)] = score(x, y)

X, Y = np.meshgrid(X, Y)

surf = ax.plot_surface(X, Y, Z, cmap=cm.jet,
                       linewidth=0, antialiased=False)

fig, ax2 = plt.subplots(1)
ax2.imshow(Z)
```
